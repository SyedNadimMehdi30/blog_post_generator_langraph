```python
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai_api_key = os.environ["OPENAI_API_KEY"]
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo")
```

```python
llm.invoke("Tell me the lastest news of USA and Iran war in 2026")
```

```python
from langchain_community.tools.tavily_search import TavilySearchResults

search = TavilySearchResults(max_results=3)
search.invoke("Tell me about USA and Iran war in 2026")
```

```python
tools = [search]
```

```python
from langgraph.prebuilt import create_react_agent 

agent_executor = create_react_agent(llm, tools)
```

```python
from langchain_core.messages import HumanMessage

response = agent_executor.invoke({"messages": [HumanMessage(content="Tell me about USA and Iran war in 2026")]})

response["messages"][3].content
```
